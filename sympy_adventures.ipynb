```python
import sympy as sp 
from sympy.plotting import plot3d

x = sp.symbols("x")
y = sp.symbols("y")
z = sp.symbols("z")
```

```python
eks = x**2 + y**2/2 + x
display(eks)
```

```python
diff_eks = sp.diff(eks,x)
display(diff_eks)
```

```python
sp.integrate(eks, (x,0,10))
```

```python
sp.limit(diff_eks,z,0)
```

```python
print(sp.solve(diff_eks))
```

```python
a = sp.Function("a")
t = sp.symbols("t")
```

```python
diff_eq = sp.Eq(a(t).diff(t,t)-a(t),5)
display(diff_eq)

sp.dsolve(diff_eq)
```
