```python
import numpy as np
import matplotlib.pyplot as plt
import collections
```

```python
kurser1 = np.array(['MekRel','MatIntro','Linalg','MatF1','Termo','MatF2','Em1'])
kar1 = np.array([2,4,7,7,7,7,4])
Middelkarakter1 = sum(kar1)/len(kar1)
Typekarakter1=collections.Counter(kar1)

kurser2 = np.array(['DatF','Optik','Kvant1','Kvant2','DynSys','StatFys','Matf3'])
kar2 = np.array([7,2,7,4,4,2,10])
Middelkarakter2 = sum(kar2)/len(kar2)
Typekarakter2=collections.Counter(kar2)

kurser3 = np.array(['Atomic','Bfilosofi','Eteori', 'EM2','VT','DidG',"Bachlor","Bachelor"])
kar3 = np.array([4,7,7,7,10,10,12,12])
Middelkarakter3 = sum(kar3)/len(kar3)
Typekarakter3=collections.Counter(kar3)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1,figsize=(11, 11))
fig.suptitle('BsC i Kvantefysik', fontsize=16)

# make a little extra space between the subplots
fig.subplots_adjust(hspace=0.5)

ax1.plot(kurser1, kar1, 'bo-')
ax1.title.set_text('1. år')
ax1.grid(True)
ax1.axhline(y=Middelkarakter1, color='b', linestyle='--', label='Gennemsnit')
ax1.axhline(y=2, color='r', linestyle='--', label = 'Dumpelinjen')

ax2.plot(kurser2, kar2, 'ro-')
ax2.set_ylabel('Karakter', fontsize=18)
ax2.title.set_text('2. år')
ax2.grid(True)
ax2.axhline(y=Middelkarakter2, color='b', linestyle='--', label='Gennemsnit')
ax2.axhline(y=2, color='r', linestyle='--', label = 'Dumpelinjen')

ax3.plot(kurser3, kar3, 'go-')
ax3.set_xlabel('Kurser', fontsize=18)
ax3.title.set_text('3. år')
ax3.grid(True)
ax3.axhline(y=Middelkarakter3, color='b', linestyle='--', label='Gennemsnit')
ax3.axhline(y=2, color='r', linestyle='--', label = 'Dumpelinjen')

plt.legend()
plt.show()

print(f'Middelkarakteren og typekarakterer på 1. år er {Middelkarakter1} og {Typekarakter1}')

print(f'Middelkarakteren og typekarakterer på 2. år er {Middelkarakter2} og {Typekarakter2}')

print(f'Middelkarakteren og typekarakterer på 3. år er {Middelkarakter3} og {Typekarakter3}')

print(f'Gennemsnit for alle 3 år er {(Middelkarakter1 + Middelkarakter2 + Middelkarakter3)/3}')
```
